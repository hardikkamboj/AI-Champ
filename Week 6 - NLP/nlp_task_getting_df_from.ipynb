# Filtering the files and combining it to a single file

```
total_df = None

for review_csv in tqdm(review_files):
  df = pd.read_csv(review_csv)
  if len(df) > 0:
    df_non_na = df.dropna()
    if type(total_df) == pd.core.frame.DataFrame:
      total_df = pd.concat([total_df,df_non_na])  
    else:
      total_df = df_non_na
```
