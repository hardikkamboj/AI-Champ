# Gates using perceptron

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# function used for plotting decision boundary 
def plot_boundary(w,b,table,title = None):
    plt.figure(figsize = (10,6))
    temp_x = np.linspace(0,1,10)
    temp_y = (-temp_x*w[0] - b) / w[1]
    sns.scatterplot('X1','X2',hue = 'y',data = table,s = 200)
    sns.lineplot(temp_x,temp_y,color = 'r',label = 'boundary')
    plt.ylim(-0.5,1.5)
    if title != None:
        plt.title(title)
```

# OR gate

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = X[:,0] | X[:,1]
```

```python
or_table = pd.DataFrame({'X1':X[:,0],'X2':X[:,1],'y':y})
or_table
```

```python
sns.scatterplot('X1','X2',hue = 'y',data = or_table,s = 200)
```

```python
# making a single layer neural network - perceptron 
or_model = Sequential(name = 'or_model')
or_model.add(Dense(1,activation = 'sigmoid',input_dim = 2))

# compiling the model 
or_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

or_model.summary()
```

```python
or_model.fit(X,y,epochs = 5000)
```

```python
np.array(or_model.weights[0])
```

```python
w,b = or_model.get_weights()
print('weights are - ',w)
print('bias is',b)
```

```python
plot_boundary(w,b,or_table,title = 'Boundary for OR GATE')
```

# And Gate

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = X[:,0] & X[:,1]

and_table = pd.DataFrame({'X1':X[:,0],'X2':X[:,1],'y':y})
and_table
```

```python
np.array(or_model.weights[0])
```

```python
w,b = and_model.get_weights()
print('weights are - ',w)
print('bias is',b)
```

```python
plot_boundary(w,b,and_table,title = 'Boundary for AND GATE')
```

# Why can't we create a similar model for XOR?

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = X[:,0] ^ X[:,1]

xor_table = pd.DataFrame({'X1':X[:,0],'X2':X[:,1],'y':y})
xor_table
```

```python
# representation of XOR
sns.scatterplot('X1','X2',hue = 'y',data = xor_table,s = 200)
```

Unlike the situations above, we can't separate the points using a single linear boundary, as to say, they are not linearily separable. Therefore we can't classify them using a single preceptron. But what we can do here is to use multi layer netword. <br> <br>
Still, lets see what happens when we use a single layer perceptron

## Using perceptron for XOR

```python
# making a single layer neural network - perceptron 
xor_model = Sequential(name = 'xor_model')
xor_model.add(Dense(1,activation = 'sigmoid',input_dim = 2))

# compiling the model 
xor_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

xor_model.summary()
```

```python
xor_model.fit(X,y,epochs= 1000)
```

```python
w,b = xor_model.get_weights()
print('weights are - ',w)
print('bias is',b)
```

```python
plot_boundary(w,b,table = xor_table,title = 'Boundary for XOR GATE')
```

So, that's how the single layer model is trying to classify XOR data, but it is just not possible to do that. And hence we are get getting accuracy of 0.5. 

## Using multi layer network

```python
xor_model_mlp = Sequential(name = 'xor_model_mlp')
xor_model_mlp.add(Dense(16,activation = 'relu',input_dim = 2))
xor_model_mlp.add(Dense(1,activation = 'softmax'))


# compiling the model 
xor_model_mlp.compile(optimizer='adam', loss='mean_squared_error', metrics=['binary_accuracy'])

xor_model_mlp.summary()
```

```python
xor_model_mlp.fit(X,y,epochs= 1000)
```
